```python
coun_text = CountVectorizer(ngram_range=(2,3), max_features=15, stop_words='english')
coun_text
```

```python
coun_df = coun_text.fit_transform(data['video_transcription_text']).toarray()
coun_df
```

```python
coun_df = pd.DataFrame(coun_df,columns=coun_text.get_feature_names_out())
coun_df.head()
```

```python
data.drop(columns=["video_transcription_text"]).reset_index(drop=True)
```

```python

data_2 = pd.concat([data.drop(columns=["video_transcription_text"]).reset_index(drop=True), coun_df], axis=1)
data_2
```

```python
data_2 = data_2.drop(columns=['#', 'video_id'])
data_2['claim_status'] = data_2['claim_status'].replace({'claim':1, "opinion":0})
```

```python
data_2.head()
```

```python
data_2 = pd.get_dummies(data_2, columns=["verified_status", "author_ban_status"], drop_first=True)
data_2.head()
```

```python
X = data_2.drop(['claim_status'], axis=1)
y = data_2['claim_status']
```

```python
X.head()
```

```python
y.head()
```

Split Data 

```python
X_tr, X_test, y_tr, y_test = train_test_split(X, y, random_state=0, test_size=0.25)
```

```python
X_train, X_val, y_train, y_val = train_test_split(X_tr, y_tr, random_state=0, test_size=0.25)
```

```python
rf = RandomForestClassifier(random_state=0)

cv_params = {'max_depth': [5, 7, None],
             'max_features': [0.3, 0.6],
            #  'max_features': 'auto'
             'max_samples': [0.7],
             'min_samples_leaf': [1,2],
             'min_samples_split': [2,3],
             'n_estimators': [75,100,200],
             }

scoring = ['accuracy', 'precision', 'recall', 'f1']

# Instantiate the GridSearchCV object
rf_cv = GridSearchCV(rf, cv_params, scoring=scoring, cv=5, refit='recall')
```

```python
rf_cv.fit(X_train, y_train)
```

```python
rf_cv.best_estimator_
```

```python
rf_cv.best_score_
```

```python
rf_cv.best_params_
```

```python
rf_cv.cv_results_
```

Evalute Metric with F1

```python
rf_cv_df = pd.DataFrame(rf_cv.cv_results_)
metric_dict = {'precision': 'mean_test_precision',
                 'recall': 'mean_test_recall',
                 'f1': 'mean_test_f1',
                 'accuracy': 'mean_test_accuracy',
                 }
rf_cv_score = rf_cv_df.iloc[rf_cv_df[metric_dict["f1"]].idxmax(), :]
accuracy = rf_cv_score[metric_dict['accuracy']]
precision = rf_cv_score[metric_dict['precision']]
recall = rf_cv_score[metric_dict['recall']]
f1 = rf_cv_score[metric_dict['f1']]
rf_cv_score_df = pd.DataFrame({
    'Model Name': ['Random Forest'],
    'precision': [precision],
    'accuracy': [accuracy],
    'recall': [recall],
    'f1': [f1]
})
```

```python
rf_cv_score_df
```

XGBOOST

```python
xgb = XGBClassifier(objective='binary:logistic', random_state=0)
cv_param = {'max_depth': [4,8,12],
             'min_child_weight': [3, 5],
             'learning_rate': [0.01, 0.1],
             'n_estimators': [300, 500]
             }

scoring = ['accuracy', 'precision', 'recall', 'f1']
xgb_cv = GridSearchCV(xgb, cv_params, scoring=scoring, cv=5, refit='recall')
```

```python
xgb_cv.fit(X_train, y_train)
```

```python
xgb_cv.best_estimator_
```

```python
xgb_cv.best_score_
```

```python
xgb_cv.best_params_
```

```python
rf_cv_pred = rf_cv.best_estimator_.predict(X_val)
```

```python
xgb_cv_pred = xgb_cv.best_estimator_.predict(X_val)
```

```python
cm = confusion_matrix(y_val, rf_cv_pred, labels=rf_cv.classes_)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=rf_cv.classes_)
disp.plot()
```

```python
cm2 = confusion_matrix(y_val, xgb_cv_pred, labels=xgb_cv.classes_)
disp = ConfusionMatrixDisplay(confusion_matrix=cm2, display_labels=xgb_cv.classes_)
disp.plot()
```

```python
target_labels = ['opinion', 'claim']
print(classification_report(y_val, rf_cv_pred, target_names=target_labels))
```

```python
target_labels = ['opinion', 'claim']
print(classification_report(y_val, xgb_cv_pred, target_names=target_labels))
```

The performance of Random Forest model better than XGBoost slighly different

Use Random forrest model to predict Test set

```python
importance = rf_cv.best_estimator_.feature_importances_
rf_df = pd.Series(importance, index=X_test.columns)
fix, ax = plt.subplots()
rf_df.plot.bar(ax=ax)
ax.set_title("Feature Importance")
ax.set_ylabel("impurity")
plt.show()
```
